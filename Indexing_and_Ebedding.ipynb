## Embedding
- Convert text into vector representations using models like OpenAI's embeddings.
- Useful for semantic search and similarity tasks.
- There are many opensource as well as priced embedding models available you ca use any based on ur usecase

### Huggingface embeddings 

```python
from langchain_huggingface.embeddings import HuggingFaceEmbeddings
```

```python
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")
```

```python
sample_text="LangChain is an amazing framework for building applications with LLMs."
```

```python
hf_embed=embeddings.embed_query(sample_text)
```

```python
type(hf_embed)
```

```python
len(hf_embed)
```

```python
hf_embed[:10]
```

### OPenai embeddings

```python
import os 
import dotenv
dotenv.load_dotenv()
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings(
    model="text-embedding-3-large",
    # With the `text-embedding-3` class
    # of models, you can specify the size
    # of the embeddings you want returned.
    # dimensions=1024
)
```

```python
openai_embed=embeddings.embed_query(sample_text)
```

```python
len(openai_embed)
```

```python
openai_embed[:10]
```
